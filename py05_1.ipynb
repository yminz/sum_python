```python
busan = pd.read_excel('부산구청.xlsx')
busan = busan.set_index('항목')
busan
```

```python
busan.loc['중구']
```

```python
import folium as f

map = f.Map(location=[35.156237, 129.022693], zoom_start=11)
# zoom : 맨 처음에 어느정도로 보이는지

map
```

```python
# 부산 지역 내의 개수 만큼 가져오기 위해 for문 이용

for item in busan.index :
    lat = busan.loc[item,'위도']
    long = busan.loc[item,'경도']
    item = item.strip()
    folium.CircleMarker([lat,long],
                       radius=df2.loc[item,'재활용처리량']/500,
                       popup=df2.loc[item,'재활용처리량'],
                       color = 'blue',
                       fill=True).add_to(map)
    
    folium.CircleMarker([lat,long],
                       radius=df2.loc[item,'음식물류발생량']/100,
                       popup=df2.loc[item,'음식물류발생량'],
                       color = 'green',
                       fill=True,).add_to(map)

 
map

# popup : 나타낼 자료
```
