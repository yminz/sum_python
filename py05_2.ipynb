## 평균 & 그룹화

```python
import pandas as pd

data = pd.read_excel('부산쓰레기발생2017년.xlsx')
```

```python
data.index
```

```python
data.columns
```

```python
c = data.columns
c = list(c)
# = list(data.colums)
c
c = [item.strip() for item in c]  
```

```python
c = [i.strip() for i in c]
c
```

```python
data.columns = c
data.columns
```

```python
data.sort_values(by='구분', ascending=False)  
```

```python
df1 = data[data['구분']=='연간(톤)']
df1
```

```python
df1 = df1.sort_values('항목')
del df1['구분']
df1
```

```python
df1 = df1.set_index('항목')
```

```python
df1 = df1.T
df1
```

```python
df1.loc['평균']=df1.mean(axis=0)
df1
```

```python
df1.loc[:,'열평균']=df1.mean(axis=1)
df1
```

```python
dfg1 = data.groupby('구분')['남구'].mean()
dfg1
```

```python
dfg = data.groupby(['구분']).mean()
dfg
```

```python
dfg2 = data.groupby(['항목']).mean()
dfg2
```
