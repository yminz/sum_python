```python
busan = pd.read_excel('부산구청.xlsx')
busan = busan.set_index('항목')
busan
```

```python
import folium as f

map = f.Map(location=[35.156237, 129.022693], zoom_start=11)
map
```

```python
for item in busan.index :
    lat = busan.loc[item,'위도']
    long = busan.loc[item,'경도']
    item = item.strip()
    f.CircleMarker([lat,long],
                       radius=dr.loc[item,'전기'],
                       popup=dr.loc[item,'전기'],
                       color = 'blue',
                       fill=True).add_to(map)
    
    f.CircleMarker([lat,long],
                       radius=dr.loc[item,'가스'],
                       popup=dr.loc[item,'가스'],
                       color = 'green',
                       fill=True,).add_to(map)
    
    f.CircleMarker([lat,long],
                       radius=dr.loc[item,'지역난방'],
                       popup=dr.loc[item,'지역난방'],
                       color = 'red',
                       fill=True,).add_to(map)

 
map
```
