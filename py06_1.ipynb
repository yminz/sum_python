```python
df = data.groupby('구명')[['전기','가스','지역난방']].mean()
df
```

```python
# index 삭제
# index로 지정되었던 것이 열 명으로 들어옴
df = df.reset_index()
df
```

```python
df_area = pd.read_excel('부산구청.xlsx')
df_area
```

```python
df_area = df_area.rename(columns = {'항목':'구명'})
df_area
```

```python
df_area['구명'].tolist()
```

```python
len(df)
```

```python
df = df.sort_values('구명')
df
```

```python
df_area = df_area.sort_values('구명')
df_area = df_area.set_index('구명')
df_area = df_area.reset_index()
df_area
```

```python
# concat : 위아래로 데이터를 병합
# axis = 1 : 양옆으로 데이터 병합
df1 = pd.concat([df,df_area])
df1
```

```python
df2 = pd.concat([df,df_area],axis=1, join='inner')
df2
```

```python
df3 = pd.merge(df, df_area, on='구명', how='inner')
df3
```

```python
import folium as f

map = f.Map(location=[35.156237, 129.022693], zoom_start=11)

for item in df3.index :
    lat = df3.loc[item,'위도']
    long = df3.loc[item,'경도']
    
    
    f.CircleMarker([lat,long],
                       radius=df3.loc[item,'전기'],
                       popup=df3.loc[item,'전기'],
                       color = 'blue',
                       fill=True).add_to(map)
    
    f.CircleMarker([lat,long],
                       radius=df3.loc[item,'가스'],
                       popup=df3.loc[item,'가스'],
                       color = 'green',
                       fill=True,).add_to(map)
    
    f.CircleMarker([lat,long],
                       radius=df3.loc[item,'지역난방'],
                       popup=df3.loc[item,'지역난방'],
                       color = 'red',
                       fill=True,).add_to(map)

 
map
```
